```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
```

```python
conn = sqlite3.connect("../databases/tech_jobs.db")

df = pd.read_sql("SELECT * FROM tech_jobs", conn)

df.head()
```

### Average Salary by Experience Level

This visualization shows the relationship between professional experience level
and average annual salary in the tech job market. It helps quantify the financial
impact of career progression over time.

```python
salary_by_exp = (
    df.groupby("experience_level")["salary_in_usd"]
    .mean()
    .sort_values()
)

salary_by_exp.plot(kind="bar")
plt.title("Avarage Salary by Experience Level")
plt.ylabel("Salary (USD)")
plt.xlabel("Experience Level")
plt.show()
```

The chart reveals a clear and consistent salary growth as experience increases.

- Entry-level (EN) roles receive the lowest average salaries, reflecting limited experience.
- Mid-level (MI) professionals show a noticeable salary increase as technical autonomy grows.
- Senior-level (SE) roles earn significantly more, indicating higher responsibility and expertise.
- Executive-level (EX) positions command the highest salaries, highlighting the strong financial
  premium associated with leadership and strategic decision-making.

This pattern reinforces the importance of long-term skill development and career progression
in data and technology-related fields.

### Top 10 Highest-Paying Job Titles

This chart highlights the job titles with the highest average salaries in the dataset,
focusing on roles that combine advanced technical expertise with leadership responsibilities.

```python
top_jobs = (
    df.groupby("job_title")["salary_in_usd"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

top_jobs.plot(kind="barh")
plt.title("Top 10 Highest-Paying Job Titles")
plt.xlabel("Avarage Salary (USD)")
plt.ylabel("Job Title")
plt.gca().invert_yaxis()
plt.show()
```

The results show that leadership and principal-level positions dominate the top salary rankings.

Roles such as Data Analytics Lead, Principal Data Engineer, and Principal Data Scientist
require deep technical knowledge combined with strategic and organizational responsibilities.
These positions often influence business decisions, system architecture, and data strategy,
which explains their higher compensation levels.

This insight is valuable for professionals aiming to plan long-term career paths in the data ecosystem.

### Remote Work Ratio and Salary Comparison

This visualization compares average salaries across different remote work arrangements
to understand how work flexibility impacts compensation in the tech industry.

```python
remote_salary = (
    df.groupby("remote_ratio")["salary_in_usd"]
    .mean()
)

remote_salary.plot(kind="bar")
plt.title("Avarage Salary by Remote Wokr Ratio")
plt.xlabel("Remote Ratio (%)")
plt.ylabel("Salary USD")
plt.show()
```

The analysis indicates that fully remote roles present the highest average salaries.

This suggests that remote work enables access to global job markets and higher-paying
opportunities, reducing geographical constraints. Hybrid roles show lower average salaries,
possibly reflecting transitional or region-limited positions.

Overall, the results reinforce the growing importance of remote work as a competitive factor
in tech careers.
