## Core Data Structures

Lists and dictionaries are fundamental for organizing
structured and semi-structured data.

```python
tech_roles = ["Data Analyst", "Data Scientist", "ML Engineer"]
tech_roles
```

```python
job_market_sample = {
    "role": "Data Scientist",
    "country": "United States",
    "avarage_salary_usd": 95000,
    "remote": True
}

job_market_sample
```

## Control Flow

Conditional logic allows data-driven decisions
based on defined rules.

```python
salary = job_market_sample["avarage_salary_usd"]

if salary >= 90000:
    salary_level = "High"
else:
    salary_level = "Medium"

salary_level
```

## Iteration and Data Exploration

Loops help explore collections of data efficiently.

```python
for role in tech_roles:
    print(f"Analyzing role: {role}")
```
