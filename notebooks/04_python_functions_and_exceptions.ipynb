# Python for Data Analysis – Functions and Error Handling

This notebook applies the concepts learned in the course
"Python para Dados – Trabalhando com funções, estruturas de dados e exceções".

The goal is to demonstrate how reusable functions and
error handling improve data analysis workflows.

```python
def classify_salary(salary_usd):
    if salary_usd >= 100000:
        return "High"
    elif salary_usd >= 70000:
        return "Medium"
    else:
        return "Low"
```

```python
classify_salary(85000)
```

## Applying Functions to Data Structures

```python
salaries = [65000, 82000, 105000, 58000]

salary_levels = [classify_salary(s) for s in salaries]
salary_levels
```

## Handling Unexpected Data

```python
def safe_salary_classification(value):
    try:
        return classify_salary(float(value))
    except (ValueError, TypeError):
        return "Invalid salary data"
```

```python
safe_salary_classification("not_a_number")
```
