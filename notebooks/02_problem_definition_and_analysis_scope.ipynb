# Problem Statement

The Brazilian technology job market is highly dynamic, with constant changes
in salary levels, demanded skills, seniority requirements, and geographic
distribution of opportunities.

Professionals, companies, and students often lack structured data-driven
insights to support career planning, hiring decisions, and market analysis.

# Analysis Objective

The objective of this project is to analyze real-world data related to the
Brazilian technology job market in order to identify trends, patterns, and
insights regarding salaries, skills demand, seniority levels, regional
differences, and remote work impact.
